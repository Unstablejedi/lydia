```python
import pandas as pd
import matplotlib.pyplot as plt 
import numpy as np 

df = pd.read_excel('Giraffe raw data.xlsx')
df
```

```python
df.isnull().count()
```

```python
df.isnull().sum()
```

```python
total_rows_filled = df.isnull().count() - df.isnull().sum()
total_rows_filled
```

```python
df.describe()
```

```python
print(df.columns.tolist())
```

```python
null_counts_by_location = df.groupby('Location').apply(lambda y:y.isnull().sum(axis=0))

print(null_counts_by_location.T)
```

```python
# Calculate null values and filled rows for each column
pd.set_option('display.max_columns', None)
null_counts_by_location_per_column = df.groupby('Location').apply(lambda y: pd.Series({
    'Total_Null_Values': y.isnull().sum(),
    'Total_Filled_Rows': len(y) - y.isnull().sum()
}))

print(null_counts_by_location_per_column)

```

```python
sorted_values = df.isnull().sum().sort_values(ascending=False)
topnullvalues = sorted_values.head(10)
topnullvalues
```

```python
top10nullcolumns = pd.DataFrame({'Column':sorted_values.head(20).index , 'Null Count':sorted_values.head(20).values})
top10nullcolumns 
```

```python
flattened_null_vals = null_counts_by_location.stack().reset_index()
flattened_null_vals.columns = ['Location','Column','Null Count']
sorted_null_values = flattened_null_vals.sort_values(by = 'Null Count',ascending=False).head(128)
sorted_null_values
```

```python
null_columns = df.isnull().sum()
```

```python
sorted_null_values.to_excel("sorted_null_values.xlsx")
null_counts_by_location.T.to_excel("null_counts_by_location.xlsx")
null_columns.to_excel("null_columns.xlsx")
null_counts_by_location_per_column.to_excel("null_counts_by_location_per_column.xlsx")
```
